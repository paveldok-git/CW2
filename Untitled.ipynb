### Выбор архитектуры нейронной сети

Это является стандартной задачей бинарной классификации, так что тут не требуется написание какой либо сложной сети, вполне достаточно многослойного перциптрона, благодаря простоте этой архитектуры, скорость и точность модели будет довольно высокой

##### MLP (модель)
Определние слоев, функции активации и количества узлов для MultiLayerPerceptron.

##### Структура МЛП
* 2 скрытых слоя
* Нормализация пакетных данных с помощью пакетной нормы между каждым слоем
* Использование функции активации ReLU между слоями
* Использование отсева перед отправкой на вывод
* Сигмоид на выходном слое, чтобы сделать вероятности от 0 до 1
* Для регуляризации будет использоваться Droput (отключение нейронов) и BatchNorm1d (нормализация пакетов)

В качестве метрик будут использованы Precision, Recall, F1 Score, эти метрики были выбраны, так как их легко интерпритировать, они быстро считаются и довольно точно определяют качество модели

#### Обоснование выбора оптимизатора:
Оптимизатор Adam является популярным выбором для обучения нейронных сетей из-за его адаптивной скорости обучения и импульса. Он сочетает в себе преимущества оптимизаторов AdaGrad и RMSProp. Адаптивная скорость обучения помогает регулировать скорость обучения для каждого параметра индивидуально, что может быть полезно при работе с различными типами данных или архитектур. Импульс помогает ускорить процесс обучения, особенно в регионах с большими градиентами. В целом Адам обеспечивает хороший баланс между скоростью сходимости и производительностью обобщения, что делает его подходящим выбором для обучения нейронных сетей.

#### Возможные модификации
Вообще для такой задачи это не совсем резонно, так как простые модели хорошо обобщают данные, если модель сильно усложнить, то и результат будет хуже, хотя один из возможных вариантов улучшения будет добавление в модель одномерной свертки, для поиска закономерностей на более глубоком уровне

### Реализация модели
#### Импорт библиотке

```python
from sklearn import preprocessing
from sklearn.metrics import *
from sklearn.metrics import f1_score as f1
from sklearn.metrics import confusion_matrix

#-- Специальные импорты для Pytorch -----#
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torch.utils.data import TensorDataset
# Отключение предупреждений (категория предупреждение = будущие обновления)
import warnings
warnings.simplefilter(action='ignore') #, category=FutureWarning)
```

#### Преобразование данных в тензоры Pytorch

```python
# Сначала используем MinMaxscaler для масштабирования всех атрибутов наборов данных Train & Test.
scaler = preprocessing.MinMaxScaler()
x_train = scaler.fit_transform(X_train.values)
x_test =  scaler.fit_transform(X_test.values)

print("Scaled values of Train set \n")
print(x_train)
print("\nScaled values of Test set \n")
print(x_test)

# Затем преобразуем наборы Train и Test в тензоры

x_tensor =  torch.from_numpy(x_train).float()
y_tensor =  torch.from_numpy(y_train.values.ravel()).float()
xtest_tensor =  torch.from_numpy(x_test).float()
ytest_tensor =  torch.from_numpy(y_test.values.ravel()).float()

print("\nTrain set Tensors \n")
print(x_tensor)
print(y_tensor)
print("\nTest set Tensors \n")
print(xtest_tensor)
print(ytest_tensor)
```

#### Настройка входных данных

DataLoader для пакетной передачи данных в модель

```python
# Определение размера пакета 
bs = 64
# И x_train, и y_train можно объединить в один TensorDataset, который будет легче перебирать и нарезать y_tensor = y_tensor.unsqueeze(1)
y_tensor = y_tensor.unsqueeze(1)
train_ds = TensorDataset(x_tensor, y_tensor)
# DataLoader от Pytorch отвечает за управление пакетами.
# Можно создать DataLoader из любого набора данных. DataLoader упрощает перебор пакетов
train_dl = DataLoader(train_ds, batch_size=bs)


# Для набора данных проверки/тестирования
ytest_tensor = ytest_tensor.unsqueeze(1)
test_ds = TensorDataset(xtest_tensor, ytest_tensor)
test_loader = DataLoader(test_ds, batch_size=64)
```

#### Реализация сети MLP

```python
# Define the neural network class
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(37, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, 2)  # Предполагая бинарную классификацию (2 выходных класса)
        self.sigmoid = nn.Sigmoid()
        
        
        # Функции активации и регурелизации
        self.relu = nn.ReLU()
        self.sigmoid =  nn.Sigmoid()
        self.dropout = nn.Dropout(p=0.1)
        self.batchnorm1 = nn.BatchNorm1d(64)
        self.batchnorm2 = nn.BatchNorm1d(64)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.batchnorm1(x)
        x = torch.relu(self.fc2(x))
        x = self.batchnorm2(x)
        x = self.dropout(x)
        x = self.sigmoid(self.fc3(x))
        return x

# Create an instance of the neural network
model = Net()
print(model)
```

```python
# Вычисление потерь
loss_func = nn.BCELoss()
# Оптимизатор
# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Function to calculate metrics
def calculate_metrics(y_true, y_pred):
    precision = precision_score(y_true, y_pred, average='weighted', labels=np.unique(y_pred))
    recall = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')
    return precision, recall, f1
num_epochs = 15
```

### Обучение и оценка качества модели

```python
train_loss = []
f1_scores = []
for epoch in range(num_epochs):
    # Training
    model.train()
    for inputs, labels in train_dl:
        optimizer.zero_grad()
        outputs = model(inputs)
        labels = labels.squeeze(1).long()  # Squeeze and convert to long data type
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    # Testing
    model.eval()
    test_loss = 0.0
    y_true = []
    y_pred = []
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            labels = labels.squeeze(1).long()  # Squeeze and convert to long data type
            batch_loss = criterion(outputs, labels)
            
            test_loss += batch_loss.item()
            _, predicted = torch.max(outputs, 1)
            y_true.extend(labels.tolist())
            y_pred.extend(predicted.tolist())
    train_loss.append(batch_loss.item())
    test_loss /= len(test_loader)
    precision, recall, f1 = calculate_metrics(y_true, y_pred)
    # Append the F1 score to the list
    f1_scores.append(f1)

    print(f"Epoch {epoch+1}:")
    print(f"Test Loss: {test_loss:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print()
    
# Plotting the F1 scores
epochs = range(1, num_epochs + 1)
plt.plot(epochs, f1_scores, marker='o')
plt.xlabel('Epoch')
plt.ylabel('F1 Score')
plt.title('F1 Score over Epochs')
plt.show()
```

Все метрики показывают 100% точность, существует 2 варианта почему так происходит:<br>

2) Простые данные - наша модель простая и хорошо умеет обобщать данные, а сами данные в свою очередь сильно разграничивают DDoS атаку и Normal поведение, так что можно сказать, что модель будет точно классифицировать DDoS и Normal категории

### Для сравнения можно обучить еще несколько сетей и сравнить их с нашей

В качестве альтернативного решения можно дороботать текщкю модель, в соответствии с планом, будет добавлена одномерная свертка

```python
# Define your neural network model
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv1d(1, 64, kernel_size=3, stride=1)
        self.conv2 = nn.Conv1d(64, 128, kernel_size=3)
        self.fc1 = nn.Linear(4224, 64)  # Adjust the input size
        self.fc2 = nn.Linear(64, 2)  # Assuming 2 output classes

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = torch.flatten(x, 1)
        x = self.fc1(x)
        x = self.fc2(x)
        return x

# Create an instance of the model
model = Net()

# Define your loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
f1_scores = []
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    total_predictions = []
    true_labels = []

    # Iterate over the training data loader
    for inputs, labels in train_dl:
        # Zero the gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(inputs.unsqueeze(1))  # Add an extra dimension for channels

        # Compute the loss
        loss = criterion(outputs, labels.squeeze().long())  # Squeeze labels to match the shape

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        # Accumulate the loss
        running_loss += loss.item()

        # Convert the outputs to predicted labels
        _, predicted = torch.max(outputs.data, 1)

        # Collect predictions and true labels for metrics calculation
        total_predictions.extend(predicted.tolist())
        true_labels.extend(labels.tolist())

    # Calculate metrics
    precision = precision_score(true_labels, total_predictions)
    recall = recall_score(true_labels, total_predictions)
    f1 = f1_score(true_labels, total_predictions)
    
    # Append the F1 score to the list
    f1_scores.append(f1)

    # Print metrics for the current epoch
    print(f"Epoch {epoch+1}/{num_epochs} - Loss: {running_loss:.4f} - Precision: {precision:.4f} - Recall: {recall:.4f} - F1 Score: {f1:.4f}")

    # Evaluation on test data
    model.eval()
    total_predictions = []
    true_labels = []

    for inputs, labels in test_loader:
        outputs = model(inputs.unsqueeze(1))  # Add an extra dimension for channels
        _, predicted = torch.max(outputs.data, 1)
        total_predictions.extend(predicted.tolist())
        true_labels.extend(labels.tolist())

    precision = precision_score(true_labels, total_predictions)
    recall = recall_score(true_labels, total_predictions)
    f1 = f1_score(true_labels, total_predictions)

    print(f"Epoch {epoch+1}:")
    print(f"Test Loss: {test_loss:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print()
    
    
# Plotting the F1 scores
epochs = range(1, num_epochs + 1)
plt.plot(epochs, f1_scores, marker='o')
plt.xlabel('Epoch')
plt.ylabel('F1 Score')
plt.title('F1 Score over Epochs')
plt.show()
```

Ради эксперемента можно в ту же изначальную сеть добавить много однотипных слоев, обычно это приводит к ухудшению показателей качества и если это подтвердится, то значит переобучения нет и модель просто хорошо обучилась

```python
# Define the neural network model
class MyModel(nn.Module):
    def __init__(self, input_size):
        super(MyModel, self).__init__()
        self.fc_layers = nn.Sequential(
            nn.Linear(input_size, 64),
            nn.Linear(64, 128),
            nn.Linear(128, 256),
            nn.Linear(256, 512),
            nn.Linear(512, 256),
            nn.Linear(256, 128),
            nn.Linear(128, 64),
            nn.Linear(64, 32),
            nn.Linear(32, 16),
            nn.Linear(16, 2)  # Assuming 2 output classes
        )

    def forward(self, x):
        x = self.fc_layers(x)
        return x

# Create an instance of the model
model = MyModel(input_size=37)

# Вычисление потерь
loss_func = nn.BCELoss()
# Оптимизатор
# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Function to calculate metrics
def calculate_metrics(y_true, y_pred):
    precision = precision_score(y_true, y_pred, average='weighted', labels=np.unique(y_pred))
    recall = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')
    return precision, recall, f1
epochs = 15

# Initialize an empty list to store F1 scores for each epoch
f1_scores = []
train_loss = []
for epoch in range(num_epochs):
    # Training
    model.train()
    for inputs, labels in train_dl:
        optimizer.zero_grad()
        outputs = model(inputs)
        labels = labels.squeeze(1).long()  # Squeeze and convert to long data type
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    # Testing
    model.eval()
    test_loss = 0.0
    y_true = []
    y_pred = []
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            labels = labels.squeeze(1).long()  # Squeeze and convert to long data type
            batch_loss = criterion(outputs, labels)
            
            test_loss += batch_loss.item()
            _, predicted = torch.max(outputs, 1)
            y_true.extend(labels.tolist())
            y_pred.extend(predicted.tolist())
    train_loss.append(batch_loss.item())
    test_loss /= len(test_loader)
    precision, recall, f1 = calculate_metrics(y_true, y_pred)
    # Append the F1 score to the list
    f1_scores.append(f1)

    print(f"Epoch {epoch+1}:")
    print(f"Test Loss: {test_loss:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print()
    
    
# Plotting the F1 scores
epochs = range(1, num_epochs + 1)
plt.plot(epochs, f1_scores, marker='o')
plt.xlabel('Epoch')
plt.ylabel('F1 Score')
plt.title('F1 Score over Epochs')
plt.show()
```
